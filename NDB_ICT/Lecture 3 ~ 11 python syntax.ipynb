## 자료형

기본적으로 프로그래밍이란 데이터를 다루는 행위이다.
따라서, 자료형에 대한 이해는 프로그래밍의 길에 있어서 첫걸음이라고 할 수 있다.

파이썬의 자료형으로는 정수형, 실수형, 복소수형, 문자열, 리스트, 튜플, 사전 등이 있다.
파이썬은 자료형이 가지고 있는 기능이 매우 다양하고 강력하다는 장점이 있다. 
가령, 파이썬의 리스트는 C++의 벡터 라이브러리 JAVA에서의 array 리스트 라이브러리가 제공하는 기본적인 기능을 이미 가지고 있어서 별도의 표준 라이브러리를 불러와 사용하지 않아도 쉽고 간결하게 기능들을 사용할 수 있다.

### 수 자료형

#### 정수형과 실수형

많은 코딩 테스트에서 정수형과 실수형이 가장 핵심적으로 사용된다.

정수형(integer)은 정수를 다루는 자료형이다.
양의 정수, 음의 정수, 0이 포함된다.

`````python
# 양의 정수
a = 1000
print(a)

# 음의 정수
a = -7
print(a)

# 0
a = 0
print(a)
`````

실수형(real number)은 소수점 아래의 데이터를 포함하는 수 자료형이다.

파이썬에서는 소수점을 붙인 수를 대입하면 실수형 변수로 처리된다.
소수부가 0이거나, 정수부가 0인 소수는 0을 생략하고 작성할 수 있다.

`````python
# 양의 실수
a = 157.93
print(a)

# 음의 실수
a = -124.25
print(a)

# 소수부가 0일 때 0을 생략
a=5.
print(a)

# 정수부가 0일 때 0을 생략
a= -.7
print(a)
`````

#### 지수 표현 방식
파이썬에서는 e나 E를 이용한 지수 표현 방식을 이용할 수 있다.
e나 E 다음에 오는 수는 10의 지수부를 의미한다.
가령, 1e9라고 입력하면, 10의 9제곱(1,000,000,000)이 된다.
$$
유효숫자 e^{지수} = 유효숫가 \times 10^{지수}
$$
지수 표현 방식은 임의의 큰 수를 표현하기 위해 자주 사용한다.

최단 경로 알고리즘에서는 도달할 수 없는 노드에 대하여 최단 거리르 무한(INF)로 설정하곤 한다.
이때 가능한 최댓값이 10억미만이라면 무한(INF)의 값으로 1e9를 이용할 수 있다.

* 문제에서 정수형 데이터를 처리한다고 하면 오류를 방지하기 위해서 내장 함수인 int()를 이용해서 정수형으로 바꾸어 출력할 수 있다.

`````python
# 1,000,000,000의 지수 표현 방식
a = 1e9
b = int(1e9)
print(a)
print(b)

# 752.5
a = 75.2e1
print(a)

# 3.853
a = 3853e-3
print(a)
`````

##### 실수형 더 알아보기

오늘날 가장 널리 쓰이는 IEEE754 표준에서는 실수형을 저장하기 위해 4바이트, 혹은 8바이트의 고정된 크기의 메모리를 할당하므로, 컴퓨터 시스템은 실수 정보를 표현하는 정확도에 한계를 가진다.

가령 10진수 체계에서는 0.3과 0.6을 더한 값이 0.9로 정확히 떨어진다. 하지만, 2진수에서는 0.9를 정확히 표현할 수 있는 방법이 없다. 이로 인해서 컴퓨터는 최대한 0.9에 가깝게 표현하지만, 미세한 오차가 발생하게 된다.

`````python
a = 0.3 + 0.6
print(a)

if a == 0.9:
    print(True)
else:
    print(False)
`````

개발 과정에서 실수값을 제대로 비교하지 못해서 원하는 결과를 얻지 못할 수 있다. 이러한 경우 round()함수를 이용할 수 있고 이러한 방법이 권장된다.

가령, 123.456을 소수 셋째 자리에서 반올림하려면 round(123.456, 2)라고 작성한다.

`````python
a = round(123.456,2)
print(a)

a = 0.3 + 0.6
print(round(a,4))

if round(a,4) == 0.9:
    print(True)
else:
    print(False)
`````

#### 수 자료형의 연산

수 자료형에 대하여 사칙연산과 나머지 연산자가 많이 사용된다.
단, 나누기 연산자(/)를 주의해야 한다.

1. 파이썬에서 나누기 연산자(/)는 나눠진 결과를 실수형으로 반환한다.
2. 나머지 연산자(%)는 나눈 나머지를 반환한다. 다양한 로직 설계에 사용하는 경우가 많다. 예) 어떤 수가 홀수인지 체크해야 하는 경우
3. 몫 연산자(//)는 몫을 반환한다.
4. 이외에도 거듭 제곱 연산자(**)를 비롯해 다양한 연산자들이 존재한다.

`````python
a = 7
b = 3

# 나누기
print(a / b)

# 나머지
print(a % b)

# 몫
print(a // b)

# 거듭 제곱
print(a ** b)

# 제곱근
print(a ** 0.5)
`````

### 리스트 자료형

리스트는 여러 개의 데이터를 연속적으로 담아 처리하기 위해 사용하는 자료형이다.

사용ㅈ 입장에서 C나 JAVA에서의 배열(array)의 기능 및 연결 리스트와 유사한 기능을 지원한다.
C++의 STL vector와 기능적으로 유사하다.
리스트 대신에 배열 혹은 테이블이라고 부르기도 한다.

#### 리스트 초기화

리스트는 대괄호([])안에 원소를 넣어 초기화하며, 쉼포(,)로 원소를 구분한다.
비어 있는 리스트를 선언하고자 할 때는 list() 혹은 간단히 []를 이용할 수 있다.
리스트의 원소에 접근할 때는 인덱스(index)값을 괄호에 넣어서 접근할 수 있으며, 인덱스는 0부터 시작한다.

`````python
# 직접 데이터를 넣어 초기화
a = [1, 2, 3, 4, 5, 6]
print(a)

# 네 번째 원소만 출력
print(a[3])

# 크기가 N=10이고, 모든 값이 0인 1차원 리스트 초기화
n = 10
a = [0]*n
print(a)
`````

#### 리스트의 인덱싱과 슬라이싱

인덱스 값을 입력하여 리스트의 특정한 원소에 접근하는 것을 인덱싱(indexing)이라고 한다.
파이썬의 인덱스 값은 양의 정수와 음의 정수를 모두 사용할 수 있다.
음의 정수를 넣는 경우 원소를 뒤에서부터 거꾸로 탐색하게 되며, 가장 뒤에 있는 원소의 인덱스는 -1부터 시작해서 앞으로갈수록 1씩 작아진다.

`````python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# 여덟 번째 원소만 출력
print(a[7])

# 뒤에서 첫 번째 원소 출력
print(a[-1])

# 뒤에서 세 번째 원소 출력
print(a[-3])

# 네 번째 원소 값 변경
a[3] = 7
print(a)
`````

리스트에서 연속적인 위치를 갖는 원소들을 가져와야 할 때는 슬라이싱(slicing)을 이용한다.
대괄호 안에 콜론(:)을 넣어서 시작 인덱스와 끝 인덱스를 설정할 수 있으며, 끝 인덱스는 실제 인덱스보다 1을 더 크게 설정해야한다.


`````python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# 두 번째 원소부터 네 번째 원소까지
print(a[1:4])
`````

#### 리스트 컴프리헨션

리스트 컴프리헨션은 리스트를 초기화하는 방법 중 하나이다. 
대괄호 안에 조건문과 반복문을 적용하여 리스트를 매우 간결하게 초기화할 수 있다.

`````python
# 0부터 0까지의 수를 포함하는 리스트
array = [i for i in range(10)]
print(array)

# 0부터 19까지의 수 중에서 홀수만 포함하는 리스트
array = [i for i in range(20) if i % 2 == 1]
print(array)

# 1부터 9까지의 수들의 제곱 값을 포함하는 리스트
array1 = [i * i for i in range(1,10)]
array2 = [i ** 2 for i in range(1,10)]
print(array1)
print(array2)
`````

#### 리스트 컴프리헨션과 일반적인 코드 비교하기

0부터 19까지의 수 중에서 홀수만 포함하는 리스트를 리스트 컴프리헨션과 일반적인 코드와 비교해보자.

`````python
# 1. 리스트 컴프리헨션
array = [i for i  in range(20) if i % 2 ==1]
print(array)

# 2. 일반적인 코드
array = []
for i in range(20):
    if i % 2 == 1:
        array.append(i)
print(array)
`````

위와 같이 여러 줄을 한줄로 줄일 수 있는 장점이 있다.

그리고 리스트 컴프리헨션은 2차원 리스트를 초기화할 때 효과적으로 사용할 수 있다. 특히, 다음과 같이 N $\times$ M 크기의 2차원 리스트를 한번에 초기화 해야 할 때 유용하다.

```python
array = [[0]*m for _in rage{n}]
```
만약 2차원 리스트를 초기화할 때 다음과 같이 작성하면 예기치 않은 결과가 나올 수 있다.

```python
array = [[0]*m]*n
```
그 이유는 python은 리스트 자료형을 이용해서 어떤 변수값을 할당하게 되면 내부적으로 그 리스트는 객체 형태로서 처리가 되고 별도의 주소값을 가지게 되기 때문이다.
그래서 단순히 길이가 m인 리스트를 n번 복사하는 것이 되기 때문에 내부 리스트가 같은 객체로 인식될 수 있다.



`````python
# N $\times$ M 크기의 2차원 리스트 초기화
n = 4
m = 3
array = [[0]*m for _ in range(n)]
print(array)

array[1][1] = 5
print(array)

# 잘못된 예시
n = 4
m = 3
array = [[0] * m] * n
print(array)

array[1][1] = 5
print(array)
`````

#### 언더바는 언제 사용하나요?

파이썬에서는 반복을 수행하되 반복을 위한 변수의 값을 무시하고자 할 때 언더바(_)를 자주 사용한다.


```python
# 코드 1. 1부터 9까지 자연수 더하기
summary = 0
for i in range(1,10):
    summary += i
print(summary)

# 코드 2. Hello World"를 5번 출력하기
for _ in range(5):
    print("Hello World")
````

위의 코드 1에는 0부터 시작해서 9까지를 변수 i에 담아서 summary에 더하는 작업에 사용하는 반면에, 코드 2에서는 range(5)로 생성되는 값들이 반복 작업에서 사용되지 않고 반복된다. 이런 경우에 언더바(_)를 사용할 수 있다.

#### 리스트에서 특정 값을 가지는 원소를 모두 제거하기

리스트에서 특정 값을 가지는 원소를 모두 제거하기 위해서 집합 자료형을 사용하면 좋다. 집합 자료형은 특정한 원소의 존재 유무만 체크하고자 할 때 효과적으로 사용할 수 있기 때문인데, 자세한 내용은 후에 다루고 지금은 특정 값을 가지는 원소를 모두 제거하는 방법에 대해서만 알아본다.

`````python
a = [1, 2, 3, 4, 5, 5, 5]
remove_set = {3, 5} # 집합 자료형

# remove_list에 포함되지 않는 값만을 저장
result = [i for i in a if i not in remove_set]
# a라는 리스트를 변수 i가 하나씩 확인하면서 i라는 변수가 remove_set에 포합되어 있지 않으면 남겨놓겠다. 
print(result)
`````

### 문자열, 튜플 자료형

문자열 변수를 초기화할 때는 큰따옴표(")나 작은 따옴표(')를 이용한다.

문자열 안에 큰 따옴표나 작은 따옴표가 포함되어야 하는 경우에 전체 문자열을 큰따옴표(작은 따옴표)를 구성하는 경우, 내부적으로 작은 따옴표(큰 따옴표)를 포함할 수 있다.
혹은, 백슬래시(\)를 사용하면, 큰따옴표나 작은 따옴표를 원하는 만큼 포함시킬 수 있다.

* 백슬래시를 이용하는 것은 이스케이프 문자를 만들기 위해서 사용할 수 있다.

`````python
data = "Hello World"
print(data)

data = "Don'y you konw \"Python\"?"
print(data)
`````

#### 문자열 연산

문자열 변수에 덧셈(+)을 이용하면 문자열이 더해져서 연결(concatenate)된다.
문자열 변수를 특정한 양의 정수와 곱하는 경우, 문자열이 그 값만큼 여러 번 더해진다.
문자열에 대해서도 인덱싱과 슬라이싱을 이용할 수 있다. 하지만, 문자열은 특정 인덱스의 값을 변경할 수는 없다.(immutable)

`````python
a = "Hello"
b = "World"
print(a + " " + b)

a = "String"
print(a * 3)

a = "ABCDEF"
print(a[2:4])
`````

`````python
a[2] = "b" # 변경 불가능
`````

#### 튜플 자료형

튜플 자료형은 리스트와 유사하지만 다음과 같은 문법적 차이가 있다.

1. 튜플은 한 번 선언된 값을 변경할 수 없다.
2. 리스트는 대괄호([])를 이용하지만, 튜플은 소괄호(())를 사용한다.

튜플은 리스트에 비해 상대적으로 공간 효율적이다. : 더 적은 메모리 사용

`````python
a = (1, 2, 3, 4, 5, 6, 7, 8, 9)

# 네 번째 원소만 출력
print(a[3])

# 두 번째 원소부터 네 번째 원소까지 출력
print(a[1:4])
`````

`````python
# 오류가 발생하는 예제
a = (1, 2, 3, 4)
print(a)

a[2] = 7
`````

튜플을 사용하면 좋은 경우

1. 서로 다른 성질의 데이터를 묶어서 관리해야 할 때
예) 최단 경로 알고리즘에서는 (비용, 노드 번호)의 형태로 튜플 자료형을 자주 사용한다.

### 사전 자료형과 집합 자료형

#### 사전 자료형

사전 자료형은 키(key)와 값(value)의 쌍을 데이터로 가지는 자료형이다.
앞서 다루었던 리스트나 튜플이 값을 순차적으로 저장하는 것과 대비된다.

사전 자료형은 키와 값의 쌍을 데이터로 가지며, 원하는 '변경 불가능한(immutable) 자료형'을 키로 사용할 수 있다.

파이썬의 사전 자료형은 해시 테이블(Hash Table)을 이용하므로 데이터의 조회 및 수정에 있어서 O(1) : 상수 시간에 처리할 수 있다는 장점이 있다. 따라서, 문자열과 같은 키를 이용해서 데이터를 관리하는 상황에는 리스트보다 빠르게 수행할 수 있다.


`````python
data = dict()
data['사과'] = 'Apple'
data['바나나'] = 'Banana'
data['코코넛'] = 'Coconut'

print(data)

data = {
    '사과' : 'Apple',
    '바나나' : 'Banana',
    '코코넛' : 'Coconut'
}

print(data)

# 특정 키의 값 출력하기
if '사과' in data:
    print("'사과'를 키로 가지는 데이터가 존재합니다. : ", data['사과'])
`````

사전 자료형에서는 키와 값을 별도로 뽑아내기 위한 메서드를 지원한다.
1. 키 데이터만 뽑아서 리스트로 이용할 때는 key()함수를 이용한다.
2. 값 데이터만 뽑아서 리스트로 이용할 때는 values()함수를 이용한다.

`````python
data = dict()
data['사과'] = 'Apple'
data['바나나'] = 'Banana'
data['코코넛'] = 'Coconut'

# 키 데이터만 담은 리스트
key_list = data.keys()
#값 데이터만 담은 리스트
value_list = data.values()
print(key_list)
print(value_list)

# 사전 키 혹은 값이라는 객체가 아닌 리스트형 데이터로 출력하기
value_list = list(data.values())
print(value_list)

# 각 키에 따른 값을 하나씩 출력
for key in key_list:
    print(data[key])
    
`````

`````python
print(data['코코넛'])
`````

#### 집합 자료형

집합은 다음과 같은 특징이 있다.
1. 중복을 허용하지 않는다.
2. 순서가 없다.
이러한 특성으로 인해서 어떤 데이터의 존재 여부만 체크할 때 유용하게 사용할 수 있다.

집합은 리스트 혹은 문자열을 이용해서 초기화할 수 있다.
set()함수를 이용하거나 중괄호({}) 안에 각 원소를 콤마(,)를 기준으로 구분하여 삽입함으로써 초기할 수 있다.

데이터의 조회 및 수정에 있어서 O(1)의 시간에 처리할 수 있다.

`````python
# 집합 자료형 초기화 방법 1
data = set([1, 1, 2, 3, 4, 4, 5])
print(data)

# 집합 자료형 초기화 방법 2
data = {1, 1, 2, 3, 4, 4, 5}
print(data)
`````

#### 집합 자료형의 연산

기본적인 집합 연산으로는 합집합, 교집합, 차집합 연산 등이 있다.
1. 합집합 : 집합 A에 속하거나 B에 속하는 원소로 이루어진 집합
2. 교집합 : 집합 A에도 속하고 B에도 속하는 원소로 이루어진 집합
3. 차집합 : 집합 A의 원소 중에서 B에 속하지 않는 원소들로 이루어진 집합

`````python
a = set([1, 2, 3, 4, 5])
b = set([3, 4, 5, 6, 7])

# 합집합
print(a | b)

# 교집합
print(a & b)

# 차집합
print(a - b)
`````

`````python
# 집합 자료형 관련 함수

data = set([1, 2, 3])
print(data)

# 새로운 원소 추가
data.add(4)
print(data)

# 새로원 원소 여러 개 추가
data.update([5, 6])
print(data)

# 특정한 값을 갖는 원소 삭제
data.remove(3)
print(data)
`````

#### 사전 자료형과 집합 자료형의 특징
1. 리스트나 튜플은 순서가 있기 때문에 인덱싱을 통해 자료형의 값을 얻을 수 있다.
2. 사전 자료형과 집합 자료형은 순서가 없기 때문에 인덱싱으로 값을 얻을 수 있다.
3. 사전의 키(key) 혹은 집합의 원소(element)를 이용해 O(1)의 시간 복잡도로 조회한다.
이때, 키나 원소의 값으로는 변경불가능한 문자열이나 튜플과 같은 객체가 사용되어야 한다.

else:
    위의 모든 조건문에 모두 Ture가 아닐 때 실행되는 코드

### 반복문

특정한 소스코드를 반복적으로 실행하고자 할 때 사용하는 문법이다.

파이썬에서는 while문과 for문이 있는데, 어떤 것을 사용해도 상관없다. 다만, 코딩 테스트에서의 실제 사용 예시를 확인해보면 for문이 더 간결한 경우가 많다.

#### while문

while문은 단순히 특정한 조건 만족할 때 반복적으로 코드 실행할 때 사용할 수 있으며, while문의 구조는 다음과 같다.

```python
while 조건문:
    조건문의 조건을 만족할 때 실행할 소스코드
```

`````python
# 1부터 9까지의 모든 정수의 합 구하기 예제

# while문
i = 1
result = 0

# i가 9보다 작거나 같을 때 아래 코드를 반복적으로 실행
while i <= 9:
    result += i
    i += 1

print(result)
`````

#### 반복문에서의 무한 루프

무한 루프(infinite loop)란 끊임없이 반복되는 반복 구문을 의미한다.
코딩 테스트에서 무한 루프를 구현할 일은 거의 없기 때문에 유의해야한다.
반복문을 작성한 뒤에는 항상 반복문을 탈출할 수 있는지 확인해야 한다.


#### for문

반복문으로 for문을 이용할 수도 있다. for문의 구조는 다음과 같은데, 특정한 변수르 이용하여 'in' 뒤에 오는 데이터(리스트, 튜플 등)에 포함되어 있는 원소를 첫 번째 인덱스부터 차례대로 하나씩 방문한다.

```python
for 변수 in 리스트:
    실행할 소스코드
```

for문에서 ㅓ연속적인 값을 차례대로 순회할 때는 range()를 주로 사용한다.
이때 range(시작 값, 끝 값 + 1) 형태로 사용한다.
인자를 하나만 넣으면 자도응로 시작값은 0이 된다.

`````python
result = 0

# i는 1부터 9까지의 모든 값을 순회
for i in range(1, 10):
    result += i
    
print(result)
`````

#### 파이썬의 반복문에서 자주 사용되는 키워드

##### continue 키워드

반복문에서 남은 코드의 실행을 건너뛰고, 다음 반복을 진행하고자 할 때 continue를 사용한다.

`````python
# 1부터 9까지의 홀수의 합을 구할 때 예제
result = 0

for i in range(1, 10):
    if i % 2 == 0:
        continue
    result += i
    
print(result)
`````

##### break 키워드

반복문을 즉시 탈출하고자 할 때 break를 사용한다.


`````python
# 1부터 5까지의 정수를 차례대로 출력하는 예제
i=1

while True:
    print("현재 i의 값 : ", i)
    if i ==5:
        break
    i += 1
`````

`````python
# 학생들의 합격 여부 판단 예제1 -점수가 80점만 넘으면 합격

scores = [95, 85, 77, 65, 97] # 앞에서부터 1,2,3,4,5번 학생
cheating_student_list = {2,4} # 컨닝한 학생 번호

for i in range(5):
    if scores[i] >=80:
        print(i+1, "번 학생은 합격입니다.") # 인덱스는 0부터 이므로
        
        
# 학생들의 합격 여부 판단 예제2 - 특정 번호의 학생은 제외하기

for i in range(5):
    if i + 1 in cheating_student_list:
        continue
    if scores[i] >= 80:
        print(i+1, "번 학생은 합격입니다.")
`````

## 함수와 람다 표현식

### 함수

함수(function)란 특정한 작업을 하나의 단위로 묶어 놓은 것을 의미한다. 이러한 작업 자체를 하나의 함수로 묶어두면 필요할 때마다 그 작업을 호출하기만 해서 전체 소스코드의 양을 줄일 수 있다. 따라서, 함수를 사용하면 불필요한 소스코드의 반복을 줄일 수 있다.

#### 함수의 종류
 
1. 내장 함수 : 파이썬이 기본적으로 제공하는 함수 예) input(), print()...
2. 사용자 정의 함수 : 개발자가 직접 정의하여 사용할 수 있는 함수


#### 함수 정의하기

프로그램에는 똑같은 코드가 반복적으로 사용되어야 할 때가 많다.
함수를 사용하면 소스코드의 길이를 줄일 수 있다.

매개변수 : 함수 내부에 사용할 변수, 함수 호출 시 사용할 입력값으로 받을 수 있다.
반환 값 : 함수에서 처리된 결과를 반환

함수의 구조는 다음과 같다.

```python
def 함수명(매개변수):
    실행할 소스코드
    return 반환 값
```

함수를 호출할 때 넣어주어야 하는 값을 인자(argument)라고 부르며, 함수를저 사용되는 매개변수는 parameter라고 한다.

`````python
# 더하기 함수 예시 1

def add(a,b):
    return a+b

print(add(3,7))
`````

`````python
# 더하기 함수 예시 2
def add(a,b):
    print("함수의 결과 : ", a + b)
    
add(3,7)
`````

 #### 파라미터 지정하기

 파라미터의 변수를 직접 지정할 수 있으며, 이 경우 매개변수의 순서가 달라도 상관없다.

`````python
# 더하기 함수 예시 2
def add(a,b):
    print("함수의 결과 : ", a + b)
    
add(a=3,b=7)
add(b=3,a=7)
`````

#### global 키워드

global 키워드로 변수를 지정하면 해당 함수에서는 지역 변수를 만들지 않고, 함수 바깥에 선언된 변수를 바로 참조하게 된다.

기본적으로 C언어 계열의 프로그래밍 언어는 global 키워드 없으면 자동을 함수 바깥에 전역변수르 참조하는데, 파이썬에서는 함수의 바깥에 정의되어 있는 변수를 사용하려면 global 키워드를 이용해야 한다.


`````python
a = 0

def func():
    global a
    a += 1
    
for i in range(10):
    func()

print(a)
`````

값을 변경하거나 새로운 값을 대입하는 것이 아니라 전역변수로 선언되어 있는 값을 그대로 출력하는 등 단순히 참조하는 경우에는 오류없이 확인할 수 있다.

전역변수로 리스트가 선언되어 있을 때, 전역변수로 선언된 리스트 객체의 내부 메서드 호출도 오류없이 수행가능하다.

함수 내부의 지역변수가 전역변수와 동일한 이름을 가지고 있다면 함수 안에서는 지역변수가 우선적으로 참조된다.

`````python
a = 0

def func_1():
    print(a)
    
func_1()

array = [1, 2, 3, 4, 5]

def func_2():
    array.append(6)
    print(array)

func_2()

def func_3():
    global array
    array=[3, 4, 5]
    array.append(6)
    print(array)

func_3()
`````

#### 여러 개의 반환 값

파이썬에서 함수는 여러 개의 반환 값을 가질 수 있다. 여러 개의 값이 한꺼번에 반환되는 것을 패킹이라고 한다. 그리고 함수를 호출하는 측면에서 반환된 값을 차례대로 특정 변수에 담는 것을 언패킹이라고 한다.

`````python
def operator(a,b):
    add_var = a + b
    subtract_var = a - b
    muliply_var = a * b
    divide_var = a / b
    return add_var, subtract_var, muliply_var, divide_var # 패킹

a, b, c, d = operator(7,3) # 언패킹
print(a, b, c, d)
`````

### 람다 표현식

람다 표현식을 이용하면 함수를 간단하게 작성할 수 있다.
특정 기능을 수행하는 함수를 한 줄에 작성할 수 있다는 점이 특징이다.

어떤 함수 자체를 입력으로 받는 또 다른 함수가 존재하는 경우 유용하게 사용할 수 있다. 또는 함수의 기능이 매우 간단하거나 한번 사용하고 마는 경우에 효과적으로 사용할 수 있다.


`````python
def add(a,b):
    return a + b

# 일반적인 add() 메서드 사용
print(add(3,7))

# 람다 표현식으로 구현한 add() 메서드
print((lambda a, b: a + b)(3,7))
`````

`````python
# 람다 표현식 예시 : 내장 함수에서 자주 사용되는 람다 함수
# 점수 순으로 오름차순 정렬
array = [('홍길동', 50), ('이순식', 32), ('아무개', 74)]

def my_key(x):
    return x[1]

print(sorted(array, key=my_key)) # 정렬 기준을 키로 리스트 정렬하기
print(sorted(array, key=lambda x: x[1])) # 람다 표현식


# 람다 표현식 예시 2 : 여러 개의 리스트에 적용
# 같은 순서의 원소끼리 합한 값을 원소로 갖는 리스트 만들기
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]

result = map(lambda a,b: a + b, list1, list2)
print(list(result))
`````
