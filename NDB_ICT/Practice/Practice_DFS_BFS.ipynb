### 내 코드의 문제점

vscode에서 f10을 누른 후 f11을 누르면 함수 내부로 들어가서 값이 어떻게 변화하는지 하나씩 볼 수 있다.

디버깅해본 결과 내 코드의 문제점은 i = 10이 되면 while문의 조건으로 끝나지만, 마지막으로 실행된 recursive_function으로 돌아가면 다시 i = 9가 되면서 while문을 다시 반복하게 된다.. 따라서, 끝나지 않는다.

이를 해결하기 위해서 return을 while문 안으로 넣어서 i = 10이 되면 이제 다시 while문을 돌지 않고 이전에 열어둔 함수들을 하나씩 닫게 된다.

````python
def recursive_function(i):
    print(f"{i}번재 함수에 의해 {i+1}번째 함수가 호출되었습니다.")
    while i < 10:
        recursive_function(i+1)
        return print(f"{i}번째 함수가 종료되었습니다.")

recursive_function(1)
````

````python
# DFS
graph = [
    [], 
    [2, 3, 8], 
    [1, 7], 
    [1, 4, 5],
    [3, 5],
    [3, 4], 
    [7],
    [2, 6, 8],
    [1, 7]
]
visitied = [False] * 9

def DFS(graph, v, visitied):
    visitied[v] = True
    print(v, end=' ')

    for i in graph[v]:
        if visitied[i] == False:
            visitied[v] = True
            DFS(graph, i, visitied)

DFS(graph, 1, visitied)
````

````python
# BFS

from collections import deque

graph = [
    [], 
    [2, 3, 8], 
    [1, 7], 
    [1, 4, 5],
    [3, 5],
    [3, 4], 
    [7],
    [2, 6, 8],
    [1, 7]
]
visitied = [False] * 9
que = deque()

def BFS(graph, v, visitied):
    que.append(v)
    visitied[v] = True
    print(v, end= ' ')
    que.popleft()

    for i in graph[v]:
        if visitied[i] == False:
            que.append(i)
            visitied[i] = True
    BFS(graph, que[0], visitied)
        


        
BFS(graph, 1, visitied)
````

````python
# BFS 재구현
from collections import deque

graph = [
    [], 
    [2, 3, 8], 
    [1, 7], 
    [1, 4, 5],
    [3, 5],
    [3, 4], 
    [7],
    [2, 6, 8],
    [1, 7]
]
visitied = [False] * 9

def BFS(graph, start, visitied):
    queue = deque([start])
    visitied[start] = True

    while queue:
        v = queue[0]
        print(v, end=' ')
        queue.popleft()

        for i in graph[v]:
            if visitied[i] == False:
                queue.append(i)
                visitied[i] = True

BFS(graph, 1, visitied)
````
