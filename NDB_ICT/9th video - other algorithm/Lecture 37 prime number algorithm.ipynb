## 1. 소수(Prime Number)

- 소수란 1보다 큰 자연수 중에서 1과 자기 자신을 제외한 자연수로는 나누어 떨어지지 않는 자연수이다.
    - 6은 1, 2, 3, 6으로 나누어떨어지므로 소수가 아니다.
    - 7은 1과 7을 제외하고는 나누어 떨어지지 않으므로 소수이다.
- 코딩 테스트에서는 어떠한 자연수가 소수인지 아닌지 판별해야 하는 문제가 자주 출제된다.

## 2. 소수의 판별 : 기본적인 알고리즘

### 구현 연습

#### 구현 계획
1. 자신보다 작은 수들로 나누었을 때 나머지가 0이 나오는 수가 있다면 소수가 아니다.
2. 나란히 붙어있는 두 수는 소수가 아니므로 두번째 전에 나오는 수까지 반복한다.
3. 짝수 판정법, 3의 배수 판정법 등 k 배수 판정법을 알고 있다면 시간은 $N - \sum_k\frac{k}{N}$만큼 줄일 수 있다.

````python
def prime(x):
    if x == 1:
        print("소수가 아닙니다.")
    else:
        i = 2
        while i < x - 1:
            if x % i == 0:
                print("소수가 아닙니다.")
                break
            i += 1
        if i == x -1 or x == 2:
            print("소수입니다.")
    
def prime_k(x):

    if x == 1:
        print("소수가 아닙니다.")
    
    elif x % 2 == 0 and x != 2:
        print("소수가 아닙니다.")
    
    else:
        i=3
        while i < x:
            if x % i == 0:
                print("소수가 아닙니다.")
                break
            i += 2
        if i == x-1 or x == 2:
            print("소수입니다.")

    
prime(2)
prime_k(2)
````

### 답안


## 3. 소수의 판별: 기본적인 알고리즘 성능 분석
- 2부터 x-1까지의 모든 자연수에 대하여 연산을 수행해야 한다.
    - 모든 수를 하나씩 확인한다는 점에서 시간 복잡도는 $O(X)$이다.

## 4. 약수의 성질
- 모든 약수가 가운데 약수를 기준으로 곱셈 연산에 대해 대칭을 이루는 것을 알수 있다.
    - 예를 들어 16의 약수는 1, 2, 4, 8, 16이다.
    - 이때 $2 \times 8 = 16$은 $8 \times 2 = 16$과 대칭이다.
- 따라서 특정한 자연수의 모든 약수를 찾을 때 가운데 약수(제곱근)까지만 확인하면 된다.
    - 예를 들어 16이 2로 나누어떨어진다는 것은 8로도 나누어떨어진다는 것을 의미한다.

## 5. 소수의 판별: 개선된 알고리즘


### 구현 연습

#### 구현 계획

1. $\sqrt(N) ^ 2 = N$이므로 약수의 개수는 $\sqrt(N)$을 기준으로 대칭이다.
2. 따라서, 나누는 과정을 $\sqrt(N)$까지만 반복한다.

#### 구현


````python
import math

def prime(x):
    mid = math.floor(math.sqrt(x))
    i = 2
    if x == 1:
        return print("소수가 아닙니다.")
    
    elif x == 2:
        return print("소수입니다.")
        
    else:
        while i <= mid:
            if x % i == 0:
                print("소수가 아닙니다.")
                break
            i += 1
        if i > mid:
            print("소수입니다.")

prime(2)

````

## 6. 소수의 판별: 개선된 알고리즘 성능 분석
- 2부터 x의 제곱근(소수점 이하 무시)까지의 모든 자연수에 대하여 연산을 수행해야 한다.
    - 시간 복잡도는 $O(\sqrt(N))$이다.

### 3. 개선된 코드 제안 (조합)

# ```python
import math

def is_prime(x):
    # 가드 절: 2 미만은 소수 아님

        return False
    
    # 2부터 루트 x까지 확인

# 함수 사용

    print(f"{num}은 소수입니다.")
else:
    print(f"{num}은 소수가 아닙니다.")
# ```

### 1. 왜 위험하다고 했을까?

#```python
# prime(3)을 실행할 때

i = 2

# while 2 <= 1 은 거짓이므로 루프를 아예 안 돕니다!
# 루프를 안 돌았으니 i는 그대로 2입니다.

### 2. "복잡하게 동작한다"의 의미

이 코드는 **상태(State) 관리가 너무 많습니다.**
* `i`라는 변수가 **'나누는 수'**도 되었다가, 루프가 끝난 뒤에는 **'소수 여부를 판별하는 기준'**으로도 쓰이죠.
* 한 변수가 두 가지 이상의 역할을 하면, 코드를 보는 사람은 `"이 `i`가 지금 무엇을 의미하는 거지?"`라고 계속 고민해야 합니다. 
* 심지어 `prime(2)`는 `elif`에서 걸러지는데, `prime(3)`은 `else`로 들어가서 루프를 돌고, `prime(4)`는 루프 중간에 `break`로 멈춥니다. **숫자마다 함수가 동작하는 경로가 너무 제각각**입니다.

### 3. 더 깔끔한 해결책: '플래그(Flag)' 변수 사용
만약 `print`를 꼭 써야 하는 상황이라면, 변수의 역할을 명확히 하는 것이 좋습니다. `i`로 소수인지 아닌지를 판단하지 말고, **"소수인가?"**라는 질문에 대한 답을 가진 변수(`is_prime`)를 따로 두는 방식입니다.

#```python
def prime(x):
    if x < 2:
        print("소수가 아닙니다.")
        return
        
    is_prime = True # 일단 소수라고 가정
    mid = math.floor(math.sqrt(x))
    
    for i in range(2, mid + 1):
        if x % i == 0:
            is_prime = False # 나누어지면 소수가 아님!
            break # 더 볼 것도 없음
            
    if is_prime:
        print("소수입니다.")
    else:
        print("소수가 아닙니다.")
#```

**이렇게 바꾸면 좋은 점:**
1. **의도가 명확함:** `is_prime`이라는 변수가 딱 소수 여부만 나타냅니다.
2. **제어 흐름이 단순함:** `for` 문은 파이썬이 알아서 돌려주고, 우리는 결과만 확인하면 됩니다. `i`가 얼마인지, `mid`와 어떤 관계인지 고민할 필요가 없습니다.

결론적으로, `print` 자체가 위험한 건 아니지만 **"로직의 결과를 변수 하나에 의존해서 if문으로 분기하는 구조"**가 커질수록 나중에 관리하기가 매우 힘들어진다는 점을 말씀드리고 싶었습니다. 어떠신가요, 이 구조가 훨씬 더 명확해 보이지 않나요?

`return`을 함수 밖에서 한다는 말은, **함수는 결과값만 '반환(Return)'하고, 그 결과를 가지고 실제로 화면에 뿌리거나(Print) 다른 작업을 하는 건 함수 호출한 쪽(밖)에서 하라**는 뜻입니다.

### 1. `print`를 안에 넣는 경우 (독립성 없음)
계산기 버튼을 눌렀는데, 계산기가 화면에 결과를 보여주는 것에서 끝나지 않고 **자기 마음대로 영수증까지 출력해서 뱉어내는 것**과 같습니다.
* **단점:** 계산 결과가 필요해서 "계산기야, 결과값만 좀 줘"라고 해도, 얘는 결과값은 안 주고 무조건 영수증(print)만 출력합니다. 결과를 다른 용도(예: 결과값에 10을 더하기)로 쓸 수가 없죠.

#```python
def check_prime(x):
    # ... 계산 로직 ...
    print("소수입니다.") # 여기서 출력을 강제함

# 함수 호출
check_prime(7) 
# 결과: 화면에 "소수입니다."가 뜸.

#```

### 2. `return`을 사용하는 경우 (독립성 확보)
계산기는 화면에 결과값만 띄워주고, **그걸 보고 영수증을 출력할지, 데이터를 저장할지는 사용자가 결정**하는 방식입니다.

#```python
def is_prime(x):
    # ... 계산 로직 ...
    return True # 결과값만 딱 던져줌

# 함수 호출 후 활용 (함수 밖에서!)
result = is_prime(7) 

if result == True:
    print("소수입니다.") # 내가 원할 때 출력함
else:
    print("소수가 아닙니다.")
#```

### 왜 이렇게 하는 게 더 좋을까? (장점)

1.  **재사용성이 극대화됩니다:** 만약 100만까지의 소수를 구해서 리스트에 담고 싶다면, `print`만 하는 함수는 쓸모가 없습니다. `return`을 사용하는 함수는 `if is_prime(i): prime_list.append(i)` 처럼 얼마든지 확장할 수 있습니다.
2.  **테스트가 쉽습니다:** 함수가 올바른 답을 내놓는지 확인하고 싶을 때, 화면에 찍히는 걸 눈으로 확인하는 대신 `if is_prime(7) == True:` 처럼 코드로 바로 검증할 수 있습니다.
3.  **역할 분리 (관심사의 분리):** 함수는 **"계산"**만 하고, 화면에 **"출력"**하는 일은 다른 코드에게 맡기는 것이 프로그래밍의 정석입니다.



**요약하면:**
함수 안에서 `print`를 써버리면 그 함수는 **"출력 전용 기계"**가 되어버립니다. `return`을 사용하면 **"데이터를 만들어내는 똑똑한 계산기"**가 됩니다.

어떤가요? 함수를 '결과를 돌려주는 기계'로 만드는 게 왜 더 유리한지 감이 좀 오시나요?

### 좋은 코드 여부

좋은 질문입니다. 성능이 비슷하다고 해서 모두 '좋은 코드'인 것은 아닙니다. 프로그래밍에서 **'좋은 코드'의 기준**은 크게 **가독성(Readability), 유지보수성(Maintainability), 그리고 확장성(Extensibility)**으로 나뉩니다.

작성하신 코드와 제가 제안한 수정안을 이 기준에서 비교해 보면, 왜 프로그래머들이 특정 스타일을 선호하는지 알 수 있습니다.

### 1. 가독성 (Readability): "의도가 명확한가?"
* **작성하신 코드:** `while` 문을 쓰고 `i`를 직접 증가(`i += 1`)시킵니다. 이는 동작 방식은 명확하지만, 파이썬에서는 `for` 문이 훨씬 직관적입니다. 코드를 읽는 사람은 "이 사람이 여기서 어떤 범위를 순회하고 싶어 하는구나"를 `range()` 구문을 통해 즉각적으로 파악할 수 있어야 합니다.
* **수정안(`for` 문 사용):** `for i in range(2, limit + 1):`이라고 적으면, 이 코드를 처음 보는 사람도 "아, 2부터 루프의 끝까지 확인하겠구나"라고 바로 이해합니다. 즉, **코드가 인간의 언어(의도)에 더 가까워집니다.**

### 2. 유지보수성 (Maintainability): "실수할 구멍이 적은가?"
* **작성하신 코드:** `i`를 루프 안에서 직접 관리(`i += 1`)할 경우, 자칫하면 `i`가 무한 루프에 빠지거나, 루프를 탈출한 후의 `i` 상태를 추적해야 하는 복잡함이 생깁니다. (예: `if i > mid:` 와 같은 조건이 뒤에 붙어야 함)
* **수정안:** `for` 문은 반복 종료 조건을 파이썬 엔진이 알아서 관리해 줍니다. 따라서 **루프 내에서 변수를 건드려 발생하는 로직 오류(Bug)가 원천 차단**됩니다. '좋은 코드'는 오류가 발생할 가능성을 구조적으로 낮춘 코드입니다.

### 3. 확장성 (Extensibility): "재사용하기 쉬운가?"
* **출력(Print) vs 반환(Return):** * 함수 내부에서 `print`를 하면, 이 함수는 **'화면에 결과를 보여주는 기능'**으로 고정됩니다. 만약 나중에 "100만 이하의 소수를 다 찾아서 리스트에 저장해라" 같은 기능을 만들고 싶어지면, 함수를 통째로 뜯어고쳐야 합니다.
    * 반대로 `return True/False`를 하면, 이 함수는 **'판별 로직'**만 담당하게 됩니다. 그러면 나중에 "소수라면 리스트에 담고, 아니라면 로그를 남겨라" 같은 다양한 기능을 함수 밖에서 쉽게 붙일 수 있습니다. 이를 **관심사의 분리**라고 합니다.
