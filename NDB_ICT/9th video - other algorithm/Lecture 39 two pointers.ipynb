## 1. 투 포인터(Two Pointers)

- 투 포인터 알고리즘은 리스트에 순차적으로 접근해야 할 때 두 개의 점의 위치를 기록하면서 처리하는 알고리즘을 의미한다. 

- 흔히 2, 3, 4, 5, 6, 7번 학생을 지목해야 할 때 간단히 '2번부터 7번까지의 학생'이라고 부르는 것처럼 리스트에 담긴 데이터에 순차적으로 접근해야 할 때는 시작점과 끝점 2개의 점으로 접급할 데이터의 범위를 표현할 수 있다.

## 2. 특정한 합을 가지는 부분 연속 수열 찾기: 문제 설명

- 투 포인터 알고리즘으로 해결할 수 있는 대표적인 문제는 특정한 합을 가지는 부분 연속 수열 찾기 문제이다.
- N개의 자연수로 구성된 수열이 있을 때, 합이 M인 부분 연속 수열의 개수를 구하는 문제이다.
- 수행 시간 제한은 O(N)이다.



## 3. 특정한 합을 가지는 부분 연속 수열 찾기: 문제 해결 아이디어

## 5. 특정합 합을 가지는 부분 연속 수열 찾기: 소스 코드

### 구현 연습

#### 구현 계획

lecture 38 참고

    2) 논리적 의사코드 작성
    3) 자료구조 및 알고리즘 선택

1. 문제 분석 및 제약 조건 파악 : 특정 합을 가지는 부분 연속 수열을 찾는데, 시간 복잡도는 O(N)이다.

2. 논리적 의사코드 작성
    1) 수열과 부분합 값을 각각 series와 m으로 받는다. 
    2) series 처음부터 시작한다.s=e=0 / counter는 0으로 초기화
    3) s+e 값과 m을 비교한다.
        s+e < m인 경우 : e += 1
        s+e = m인 경우 : counter +=1, s+=1
        s+e > m인 경우 : s += 1
    4) s가 series 끝에 도착하면 알고리즘을 끝낸다.

3. 자료구조 및 알고리즘 선택
    1) series = 리스트
    2) m, counter = int
    3) 핵심 알고리즘 반복문은 while 문으로 조건은 s <>= len(series)

````python
# series = list(map(int, input().split()))
# m = int(input())
series = [1, 2, 3, 2, 5]
m = 5

counter = 0
s=0
e=0
result = 0

while len(series)-1 >= s:
    result = sum(series[s:e+1])
    if result < m:
        e += 1
    elif result == m:
        counter += 1
        s+=1
    else:
        s += 1

print(counter )
````

#### 구현 중 오류
1. 리스트의 s부터 e값까지 더해야 하는데, s+e로 해놔서 series[s]+series[e]로 갔다가 슬라이싱해서 더하는 것으로 수정함.

n = 5 # 데이터의 개수 N
m = 5 # 찾고자 하는 부분합 M
data = [1, 2, 3, 2, 5] # 전체 수열

count = 0
interval_sum = 0
end = 0

# start를 차례대로 증가시키며 반복
for start in range(n):
    # end를 가능한 만큼 이동시키기
    while interval_sim < m and end < n:
        interval_sum += data[end]
        end += 1
    # 부분합이 m일 때 카운트 증가
    if interval_sum == m:
        count += 1
    interval_sum -= data[start]

print(count)

### 1. 피드백: 왜 본인의 코드는 느린가요?

가장 결정적인 차이는 **`sum(series[s:e+1])`** 부분입니다.

* **현재 코드의 문제점:** `while` 루프를 돌 때마다 `sum()` 함수를 호출하여 리스트의 일부 구간을 처음부터 끝까지 다시 더하고 있습니다. 리스트의 길이를 $N$이라고 할 때, `sum()` 연산은 그 자체로 $O(N)$의 시간이 걸립니다. 이를 루프 안에서 반복하면 **최종 시간 복잡도는 $O(N^2)$**이 됩니다.
* **답안 코드의 장점:** 답안은 `interval_sum`이라는 변수를 사용하여, `end`가 움직일 때는 값을 더하고, `start`가 움직일 때는 값을 빼는 방식으로 **현재 구간의 합을 항상 유지**합니다. 이렇게 하면 매번 새로 더할 필요가 없으므로 **시간 복잡도가 $O(N)$**이 됩니다.

### 2. 코드 개선 제안

작성하신 코드를 $O(N)$으로 개선하려면 다음과 같이 수정할 수 있습니다.

#```python
series = [1, 2, 3, 2, 5]
m = 5

counter = 0
s = 0
current_sum = 0 # sum() 대신 현재 합을 저장하는 변수 사용

for e in range(len(series)):
    # 1. 구간에 새로운 값을 추가
    current_sum += series[e]
    
    # 2. 합이 m보다 크면, s를 오른쪽으로 옮기면서 값을 뺌
    while current_sum > m and s <= e:
        current_sum -= series[s]
        s += 1
    
    # 3. 합이 m과 같으면 카운트 증가
    if current_sum == m:
        counter += 1

print(counter)
#```



### 3. 논리적 비교 및 조언

| 구분 | 본인의 풀이 | 개선된 풀이 (답안 방식) |
| :--- | :--- | :--- |
| **핵심 연산** | `sum(series[s:e+1])` | `current_sum +=`, `-=` |
| **시간 복잡도** | $O(N^2)$ | $O(N)$ |
| **가독성** | 직관적이나 비효율적 | 알고리즘의 표준 패턴 |

**작성하신 의사코드에 대한 평가:**
논리는 훌륭합니다. 특히 `s+e`라고 표현하신 부분(의도하신 것은 `s`부터 `e`까지의 구간 합)을 `interval_sum` 변수로 관리해야겠다는 생각만 연결하시면 완벽합니다.

**추천 학습 방향:**
이 알고리즘을 **"슬라이딩 윈도우(Sliding Window)"** 기법이라고 합니다. 창문(window)이 오른쪽으로 밀려가듯, 구간의 끝을 확장하고 시작을 축소하며 합을 조절하는 개념입니다. 이 개념을 이해하고 나면, 이후에 나올 더 복잡한 배열 문제들도 쉽게 해결하실 수 있을 거예요.
