## 1. 트리(Tree)

- 트리는 가계도와 같은 계층적인 구조를 표현할 때 사용할 수 있는 자료구조이다.
- 보통 트리 구조를 표현할 때는 뿌리가 위쪽으로 올 수 있도록 한다.
- 아래의 에시에서는 A가 뿌리인 루트 노드이다.



>[트리 관련 용어]
- 루트 노드(root node) : 부모가 없는 최상위 노드
- 단말 노드(leaf node) : 자식이 없는 노드
- 크기(size) : 트리에 포함된 모든 노드의 개수
- 깊이(depth) : 루트 노드부터의 거리
- 높이(height) : 깊이 중 최댓값
- 차수(degree) : 각 노드의 (자식 방향) 간선의 개수

- 기본적으로 트리의 크기가 N일 때, 전체 간선의 개수는 N-1개이다.

## 2. 이진 탐색 트리(Binary Search Tree)

- 이진 탐색이 동작할 수 있도록 고안딘 효율적인 탐색이 가능한 자료구조의 일종이다.
- 이진 탐색 트리의 특징 : 왼쪽 자식 노드 < 부모 노드 < 오른쪽 자식 노드
    - 부모 노드보다 왼쪽 자식 노드가 작다.
    - 부모 노드보다 오른쪽 자식 노드가 크다.
* 여기서는 이진 탐색 트리를 어덯게 만들 수 있는지는 생략하고 이진 탐색 트리에서 특정 데이터 찾는 방법에 대해서 소개한다.

## 3. 트리의 순회 (Tree Traversal)

- 트리의 순회란 트리 자료구조에 포함된 노드를 특정한 방법으로 한 번씩 방문하는 방법을 의미한다.
    - 트리의 정보를 시각적으로 확인할 수 있다. 이러한 이유로 트리 자료구조를 실제로 구현할 때 트리의 순회도 같이 구현해서 사용하는 경우가 많다.
- 대표적인 트리 순회 방법은 다음과 같다.
    - 전위 순회(pre-order traverse) : 루트를 먼저 방문한다.
    - 중위 순회(in-order traverse) : 왼쪽 자식을 방문한 뒤에 루트를 방문한다.
    - 후위 순회(post-order traverse) : 오른쪽 자식을 방문한 뒤에 루트를 방문한다.
