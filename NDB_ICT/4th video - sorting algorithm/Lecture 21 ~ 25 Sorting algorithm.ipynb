### 1. 정렬 알고리즘

정렬(Sorting)이란 데이터를 특정한 기준에 다라 순서대로 나열하는 것을 말한다.
일반적으로 문제 상황에 따라서 적절한 정렬 알고리즘이 공식처럼 사용된다.
* 예) 데이터 개수가 적을 때, 데이터 개수가 많지만 범위가 특정 범위로 한저오디어 있을 대, 데이터가 거의 정렬되어 있을 때 등... 이럴 때 적절한 정렬 알고리즘이 공식처럼 사용된다.



위와 같은 상황에서 사람은 '0~9까지 하나씩 있는 것을 보니까 내림차순 혹은 오름차순으로 정렬하면 되겠다'고 알 수 있다. 하지만, 컴퓨터는 정렬 방법을 구체적인 알고리즘으로 정확히 명시해주지 않으면 정렬을 수행하기 어렵다. 따라서, 정렬을 수행하는 프로그램을 작성할 때는 정확히 어떤 방식으로 정렬을 수행할 수 있을지 알고리즘을 코드를 이용해서 정확히 표현해야 한다.

* 이 강의에서는 오름차순을 기준으로 한다.

### 2. 선택 정렬

처리되지 않은 데이터 중에서 가장 작은 데이터를 선택해 맨 앞에 있는 데이터와 바꾸는 것을 반복한다.

### 3. 선택 정렬 동작 예시










탐색 범위는 갈수록 줄어들고 매번 탐색 범위만큼 작은 원소 찾아야 하므로 선형 탐색 반복 -> 이중 반복문을 구현 가능

### 4. 선택 정렬 소스코드

````python
# 구현 연습
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(len(array)):
    min_index = i 
    for j in range(i+1, len(array)):
        if array[min_index] > array[j]:
            min_index = j
    array[i], array[min_index] = array[min_index], array[i] 
print(array)
````

### 5. 선택 정렬의 시간 복잡도

선택 정렬은 N번 만큼 가장 작은 수를 찾아서 맨 앞으로 보내야 한다.

구현 방식에 따라서 사소한 오차는 있을 수 있지만, 전체 연산 횟수는 다음과 같다.
$$
N + (N-1) + (N-2) + ... + 2
$$

이는 $\frac{(N^2 + N - 2)}{2}$로 표현할 수 있는데, 빅오 표기법에 따라서 $O(N^2)$으로 표현할 수 있다.

### 1. 퀵 정렬

퀵 정렬은 기준 데이터를 설정하고 그 기준보다 큰 데이터와 작은 데이터의 위치를 바꾸는 방법이다.
일반적인 상황에서 데이터의 특성과 관련없이 가장 많이 사용되는 정렬 알고리즘 중 하나이다.
병합 정렬과 더불어 대부분의 프로그래밍 언어의 정렬 라이브러리의 근간이 되는 알고리즘이다.
* C, JAVA, Python의 표준 정렬 라이브러리 모두 퀵 정렬 혹은병합 정렬의 아이디어를 채택한 하이브리드 방식의 정렬 알고리즘을 사용하고 있다.
가장 기본적인 퀵 정렬은 첫 번째 데이터를 기준 데이터(Pivot)로 설정한다.

### 2. 퀵 정렬 동작 예시







* 분할을 Partition이라고도 부른다.




왼쪽과 오른쪽에 있는 데이터를 각각 배열로 생각해서 퀵 정렬을 다시 수행해준다.






퀵 정렬은 재귀적으로 작동하며, 수행 범위가 점점 작아진다.


### 3. 퀵 정렬이 빠른 이유



#### 시간 복잡도 $O(\log N)$

데이터 N개에서 연산을 통해서 개수가 절반씩 줄어든다면 N개의 데이터를 계산하기 위해 필요한 연산 횟수는 다음과 같다.

| 연산 수행 횟수 | 데이터 개수 | 
|:--:|:--:|
| 0번(최소) | $N$ 개 |
| 1번 | $\frac{N}{2}$ 개 | 
| 2번 | $\frac{N}{2^2}$ 개 |
| 3번 | $\frac{N}{2^3}$ 개 |

데이터가 1개가 되기 위해서 필요한 연산의 횟수를 $k$라고 하면 다음의 식으로 부터 구할 수 있다.

$$
\frac{N}{2^k} = 1 \Rightarrow N = 2^k \Rightarrow k = \log_{2} N 
$$


### 4. 퀵 정렬의 시간 복잡도

퀵 정렬은 평균의 경우 $O(N \log N)$의 시간 복잡도를 갖는다.
* 이는 수학적으로 증명이 가능하다.
최악의 경우에는 $O(N^2)$의 시간 복잡도를 갖는다.
* 최악의 경우는 피벗의 설정으로 인해 분할이 절반에 가깝지 않고 한쪽으로 편향된 분할이 발생한 경우이다.
* 예)첫 번째 원소를 피벗으로 삼고 이미 정렬된 배열에 대해서 퀵 정렬을 수행하는 경우이다.
1. 0 ~ 9 까지 정렬된 배열에 대해서 0을 피벗으로 삼으면 큰 수 1을 바로 찾지만, 작은 수는 없으므로 0이 나올 때까지 비교한다. 이후 0을 제외한 배열을 오른쪽 배열로 나눠서 진행한다. - N
2. 1 ~ 9까지도 모두 정렬된 배열에 대해서 1을 피벗으로 삼기 때문에 바로 옆에 큰 수 2를 바로 찾지만, 작은 수 없으므로 1이 나올 때까지 비교 한다. 이후 1을 제외한 배열을 오른쪽 배열로 나눠서 진행한다. N-1
3. 이것을 반복하면 시간 복잡도는 $O(N^2)$이 된다.

* 다양한 프로그래밍 언어에서 표준 정렬 라이브러리를 적용할 때, 이 퀵 정렬을 기반으로 라이브러리가 작성되어 있는 경우 최악의 경우에도 $O(N \log N)$을 보장하는 형태로 구현한다.

### 5. 퀵 정렬 소스코드 : 일반적인 방식


````python
# 구현 연습 

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick(array, pivot):
    
    for i in range(len(array)):
        if array[i] >= array[pivot]:
            break
    
    for j in range(len(array), 0, -1):
        if array[j] <= array[pivot]:
            break
    if i < j:
        array[i], array[j] = array[j], array[j]
    
    else:
        array[j], array[pivot] = array[pivot], array[j]
        quick(array[:pivot])
        quick(array[pivot:])
        return array

quick(array, 0)
        
````

#### 나의 코드가 틀린 이유

1. 재귀 함수 구현 중 인자를 전부 주지 않았음.
2. 너무 복잡하게 생각했음.

#### 정답

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    if start >= end: # 원소가 1개인 경우 종료
        rerturn
    pivot = start # 피벗은 첫 번재 원소
    left = start + 1
    right = end
    while(left <= right):
        # 피벗보다 큰 데이터를 찾을 때까지 반복
        while(left <= end and array[left] <= array[pivot]):
            left += 1
        # 피벗보다 작은 데이터를 찾을 때까지 반복
        while(right > start and array[right] >= array[pivot]):
            right -= 1
        if(left > right): # 엇갈렸다면 작은 데이터와 피벗을 교체
            array[right], array[pivot] = array[pivot], array[right]
        else: # 엇갈리지 않았다면 작은 데이터와 큰 데이터를 교체
            array[left], array[right] = array[right], array[left]
    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행
    quick_sort(array, start, right - 1)
    quick_sort(array, right + 1, end)

quick_sort(array, 0, len(array) - 1)
print(array)
```

### 6. 퀵 정렬 소스코드 : 파이썬의 장점을 살린 방식

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    # 리스트가 하나 이하의 원소만 담고 있다면 종료
    if len(array) <= 1:
        return array
    pivot = array[0] # 피벗은 첫 번째 원소
    tail = array[1:] # 피벗을 제외한 리스트

    left_side = [x for x in tail if x <= pivot] # 분할된 왼쪽 부분
    right_side = [x for x in tail if x > picot] # 분할된 오른쪽 부분

    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행하고, 전체 리스트 반환
    return quick_sort(left_side) + [pivot] + quick_sort(right_side)

print(quick_sort(array))
```

### 1. 계수 정렬

계수 정렬 알고리즘은 특정한 조건이 부합할 때만 사용할 수 있지만, 매우 빠르게 동작하는 정렬 알고리즘이다.
* 계수 정렬은 데이터의 크기 범위가 제한되어 정수 형태로 표현할 수 있을 때 사용 가능하다.

데이터의 개수가 N, 데이터(양수) 중 최댓값이 K일 때 최악의 경우에도 수행시간 $O(N + K)$를 보장한다.

### 2. 계수 정렬 동작 예시










각 데이터 몇 번 등장했는지 세는 아록리즘 모든 범위를 포함하는 배열을 만들어야 해서 상대적으로 공간 복잡도가 높지만, 퀵 정렬과 비교했을 때 조건만 만족하면 더 빠르게 동작할 수 있다. 

### 3. 계수 정렬 소스코드

````python
# 구현 연습

array = [7, 5, 9, 0, 3, 1, 6, 2, 9, 1, 4, 8, 0, 5, 2]

count = [0]*10
for k in array:
    count[k] += 1

array = []
for x in range(len(count)):
        for i in range(count[x]):
            array.append(x)
        
print(array)
````

### 4. 계수 정렬의 복잡도 분석

1. 계수 정렬의 시간 복잡도와 공간 복잡도는 모두 $O(N + K)$이다.
2. 계수 정렬은 때에 따라서 심각한 비효율성을 초래할 수 있다.

* 예) 성적 100점을 맞은 학생이 여러 명인 경우

### 1. 정렬 알고리즘 비교하기

앞서 배운 네 가지 정렬 알고리즘을 정리하면 다음과 같다.
추가적으로, 대부분의 프로그래밍 언어에서 지원하는 표준 정렬 라이브러리는 최악의 겨웅에도 $O(N \log N)$을 보장하도록 설계되어 있다.

|정렬 알고리즘| 평균 시간 복잡도| 공간 복잡도 | 특징|
|:--:|:--:|:--:|:--:|
|선택 정렬| $O(N^2)$|$O(N)$|아이디어가 매우 간단하다.|
|삽입 정렬| $O(N^2)$|$O(N)$|데이터가 거의 정려뢰어 있을 때 가장 빠르다.|
|퀵 정렬|$O(N \log N)$|$O(N)$|대부부의 경우에 가장 적합하며, 충분히 빠르다.|
|계수 정렬| $O(N+K)$|$O(N+K)$|데이터의 크기가 한정되어 있는 경우에만 사용이 가능하지만 매우 빠르게 동작한다.|



#### 선택 정렬과 기본 정렬 라이브러리 수행 시간 비교



````python
# 구현 연습
# 선택 정렬과 기본 정렬 라이브러리의 수행 시간을 비교한다. 
# 구현 연습해야 할 부분은 선택 정렬 구현하기

from random import randint
import time

# 배열에 10,000r개의 정수를 삽입
array = []
for _ in range(10000):
    # 1부터 100사이의 랜덤한 정수
    array.append(randint(1,100))

start_time = time.time()

# 선택 정렬 프로그램 소스코드 
# [To Do]
for i in range(len(array)):
    min = i
    for j in range(i+1, len(array)):
        if array[min] > array[j]:
            array[min], array[j] = array[j], array[min]

# 측정 종료
end_time = time.time()
# 수행 시간 출력
print("선택 정렬 성능 측정:", end_time - start_time)

# 배열을 다시 무작위 데이터로 초기화
array = []
for _ in range(10000):
    # 1부터 100 사이의 랜덤한 정수
    array.append(randint(1,100))

# 기본 정렬 라이브러리 성능 측정
start_time = time.time()

# 기본 정렬 라이브러리 사용
array.sort()

# 측정 종료
end_time = time.time()
# 수행 시간 출력
print("기본 정렬 라이브러리 성능 측정:", end_time - start_time)
````

### 2. 기초 문제 풀이

#### <문제> 두 배열의 원소 교체

##### 문제 설명

동빈이는 두 개의 배열 A와 B를 가지고 있스빈다. 두 배열은 N개의 원소로 구성되어 있으며, 배열의 원소는 모두 자연수입니다.
동빈이는 최대 K번의 바꿔치기 연산을 수행할 수 있는데, 바꿔치기 연산이란 배열 A에 있는 원소 하나와 배열 B에 있는 원소 하나를 골라서 두 원소를 서로 바꾸는 것을 말합ㄴ디ㅏ.
동빈이의 최종 목표는 배열 A의 모든 원소의 합이 최대가 되도록 하는 것이며, 여러분은 동빈이를 도와야 합니다.
N, K 그리고 배열 A와 B의 정보가 주어졌을 때, 최대 K 번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 A의 모든 원소의 합을 최대값을 출력하는 프로그램을 작성하세요.


에를 들어 N = 5, K = 3이고, 배열 A와 B가 다음과 같다고 해봅시다.
배열 A = [1, 2, 3, 4, 5]
배열 B = [5, 5, 6, 6, 5]

이 경우, 다음과 같이 세 번의 연산을 수행할 수 있습니다.
연산 1) 배열 A의 원소 '1'과 배열 B의 원소 '6'을 바꾸기
연산 2) 배열 A의 원소 '2'와 배열 B의 원소 '6'을 바꾸기
연산 3) 배열 A의 원소 '3'과 배열 B의 원소 '5'를 바꾸기

세 번의 연산 이후 배열 A와 배열 B의 상태는 다음과 같이 구성될 것입니다.
배열 A = [6, 6, 5, 4, 5]
배열 B = [3, 5, 1, 2, 5]

이때 배열 A의 모든 원소의 합은 26이 되며, 이보다 더 합을 크게 만들 수는 없습니다.

난이도 하 | 풀이 시간 15분 | 시간 제한 2초 | 메모리 제한 128MB

입력 조건
1. 첫 번째 줄에 N, K가 공백을 기준으로 구분되어 입력됩니다. ( 1 <= N <= 100,000, 0 <= K <= N)
2. 두 번째 줄에 배열 A의 원소들이 공백을 기준으로 구분되어 입력됩니다. 모든 원소는 10,000,000보다 작은 자연수입니다.
3. 세 번째 줄에 배열 B의 원소들이 공백을 기준으로 구분되어 입력됩니다. 모든 원소는 10,000,000보다 작은 자연수 입니다.

출력 조건
1. 최대 K번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 A의 모든 원소의 합의 최댓값을 출력합니다.

입력 예시
```
5 3
1 2 5 4 3
5 5 6 6 5
```
출력 예시
```
26
```

##### 나의 풀이

문제 풀이 전략

1. 배열 A의 원소 중 가장 작은 원소를 찾는다. 
2. 배열 B의 원소 중 가장 큰 원소를 찾는다.
3. 1번과 2번의 결과를 비교 했을 때, 배열 A의 원소가 더 큰 경우에 바꿔치기 연산을 수행한다.
4. 위의 과정을 K번 혹은 3번에서 바꿔치기 연산이 더 이상 진행되지 않을 때까지 진행한다.


````python
# 입력
N, K  = map(int, input().split(' '))
A = list(map(int, input().split(' ')))
B = list(map(int, input().split(' ')))

# 정렬
for k in range(K):
    min = 0
    for i in range(N):
        if A[min] > A[i]:
            min = i

    max = 0
    for j in range(N):
        if B[max] < B[j]:
            max = j

    if A[min] < B[max]:
        A[min], B[max] = B[max], A[min]
    else:
        break

print(sum(A))
````

##### 저자 풀이

###### 문제 해결 아이디어

핵심 아이디어 : 매번 배열 A에서 가장 작은 원소를 골라서, 배열 B에서 가장 큰 원소와 교체합니다.
가장 먼저 배열 A와 B가 주어지면 A에 대하여 오름차순 정렬하고, B에 대하여 내림차순 정렬합니다.
이후에 두 배열의 원소를 첫 번재 인덱스부터 차례로 확인하면서 A의 원소가 B의 원소보다 작을 때에만 교체를 수행합니다.
이 문제에서는 두 배열의 원소가 최대 100,000개까지 입력될 수 있으므로, 최악의 경우$O(N \log N)$을 보장하는 정렬 알고리즘을 이용해야 합니다.

###### 답안 예시

```python
n, k = map(int, input.split())# N과 K를 입력 받기
a = lsit(map(int, input().split())) # 배열 A의 모든 원소를 입력 받기
b = lsit(map(int, input().split())) # 배열 B의 모든 원소를 입력 받기

a.sort() # 배열 A는 오름차순 정렬 수행
b.sort(reverse=True) # 배열 B는 내림차순 정렬 수행

# 첫 번째 인덱스부터 확인하며, 두 배열의 원소를 최대 K번 비교
for i in ragne(k):
    # A의 원소가 B의 원소보다 작은 경우
    if a[i] < b[i] :
        # 두 원소를 교체
        a[i], b[i] = b[i], a[i]
    else: # A의 원소가 B의 원소보다 크거나 같을 때, 반복문을 탈출
        break

print(sum(a)) # 배열 A의 모든 원소의 합을 출력
```

##### 나의 부족한 점

나의 코드에서는 시간 복잡도를 고려하지 않았다.

나의 코드의 시간 복잡도는 배열 A와 배열 B에서 가장 작은 원소를 찾는데에 $O(N)$ 그리고 최대 K번 반복하므로 최종 시간 복잡도는 $O(K \times N)$이 된다.

그러나, sort를 이용한다면, 시간 복잡도는 $O(N \log N + K)$로 최종 시간 복잡도는 $O(N \log N)$으로 나의 코드보다 더 빠르다.
